```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.api import SimpleExpSmoothing
from statsmodels.tsa.filters.hp_filter import hpfilter
import matplotlib.gridspec as gridspec

# -------------------------------
# 1. Load Data
# -------------------------------
df = pd.read_csv("StockData_Samsung_final.csv")
df["Date"] = pd.to_datetime(df["Date"])
ts = df["Adj Close"].dropna()
dates = df["Date"]

# -------------------------------
# 2. Stationarity Test Function
# -------------------------------
def test_stationarity(ts_data):
    ts_data = ts_data.dropna()
    results = {}

    adf_result = adfuller(ts_data, autolag="AIC")
    kpss_result = kpss(ts_data, regression='c', nlags='auto')

    results["adf_p"] = adf_result[1]
    results["adf_stationary"] = adf_result[1] < 0.05
    results["kpss_p"] = kpss_result[1]
    results["kpss_stationary"] = kpss_result[1] > 0.05

    return results

# -------------------------------
# 3. Define Transformations
# -------------------------------
transformations = {}

transformations["1. Differenz"] = ts.diff()
transformations["2. Differenz"] = ts.diff().diff()
log_ts = np.log(ts)
transformations["Log"] = log_ts
transformations["Log-Differenz"] = log_ts.diff()
rolling_mean = ts.rolling(window=5).mean()
transformations["Moving Average (Residuum)"] = ts - rolling_mean
ses_model = SimpleExpSmoothing(ts, initialization_method="heuristic").fit(smoothing_level=0.2, optimized=False)
ses_fitted = ses_model.fittedvalues
transformations["Exp. Smoothing (Residuum)"] = ts - ses_fitted
cycle, trend = hpfilter(ts, lamb=1600)
transformations["HP-Filter (Zyklus)"] = cycle

# -------------------------------
# 4. Plot All
# -------------------------------
plt.style.use("seaborn-whitegrid")
fig = plt.figure(figsize=(18, 12))
gs = gridspec.GridSpec(3, 3, height_ratios=[1.2, 1, 1])
fig.suptitle("Zeitreihen-Transformationen mit Stationaritätstests", fontsize=18)

# Plot Original Time Series
ax0 = plt.subplot(gs[0, :])
ax0.plot(dates, ts, color="steelblue")
ax0.set_title("Originale Zeitreihe", fontsize=14)
ax0.grid(True)

orig_results = test_stationarity(ts)
textstr = f"ADF p = {orig_results['adf_p']:.3f} → {'Stationär' if orig_results['adf_stationary'] else 'Nicht Stationär'}\n" \
          f"KPSS p = {orig_results['kpss_p']:.3f} → {'Stationär' if orig_results['kpss_stationary'] else 'Nicht Stationär'}"
ax0.text(0.02, 0.85, textstr, transform=ax0.transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.9))

# Plot 6 Transformations in 2x3 Grid
plot_titles = list(transformations.keys())
for i, title in enumerate(plot_titles[:6]):  # Only show 6 transformations
    row, col = divmod(i, 3)
    ax = plt.subplot(gs[row + 1, col])
    series = transformations[title].dropna()
    aligned_dates = dates[-len(series):]  # Align with length of cleaned series

    ax.plot(aligned_dates, series, color="#4682B4")
    ax.set_title(title, fontsize=12)
    ax.tick_params(axis='x', rotation=30)
    ax.grid(True)

    results = test_stationarity(series)
    test_text = f"ADF p = {results['adf_p']:.3f} → {'Stationär' if results['adf_stationary'] else 'Nicht Stationär'}\n" \
                f"KPSS p = {results['kpss_p']:.3f} → {'Stationär' if results['kpss_stationary'] else 'Nicht Stationär'}"
    ax.text(0.02, 0.85, test_text, transform=ax.transAxes, fontsize=9,
            verticalalignment='top', bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.9))

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()


from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Score function: the lower the score, the better the stationarity
def compute_stationarity_score(series):
    series = series.dropna()
    adf_p = adfuller(series, autolag="AIC")[1]
    kpss_p = kpss(series, regression='c', nlags='auto')[1]
    # We want low ADF p and high KPSS p → invert KPSS
    score = adf_p + (1 - kpss_p)
    return score

# Evaluate all transformed series and find the best
best_name = None
best_series = None
best_score = float('inf')

for name, series in transformations.items():
    try:
        score = compute_stationarity_score(series)
        if score < best_score:
            best_score = score
            best_name = name
            best_series = series.dropna()
    except Exception as e:
        print(f"Fehler bei {name}: {e}")

print(f"\n👉 Beste stationäre Transformation: {best_name} (Score = {best_score:.3f})")

# Plot ACF and PACF of best series
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f"ACF & PACF für '{best_name}'", fontsize=16)

plot_acf(best_series, ax=axes[0], lags=40)
axes[0].set_title("Autokorrelationsfunktion (ACF)")

plot_pacf(best_series, ax=axes[1], lags=40, method='ywm')
axes[1].set_title("Partielle Autokorrelationsfunktion (PACF)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pmdarima.arima import auto_arima
from statsmodels.tsa.arima.model import ARIMA

def compare_forecasts(series, original_series, full_log_series,
                      splits=[0.2, 0.4, 0.6, 0.8], steps_fraction=0.2,
                      arima_order=(1, 0, 1), future_fraction=0.15):
    n = len(series)
    total_plots = len(splits) + 1  # Including the future forecast
    fig, axs = plt.subplots(total_plots, 1, figsize=(14, 4 * total_plots), sharex=True)

    # Loop through backtesting splits
    for i, split in enumerate(splits):
        train_end = int(split * n)
        test_end = int((split + steps_fraction) * n)
        train, test = series[:train_end], series[train_end:test_end]

        # Auto ARIMA
        auto_model = auto_arima(train, seasonal=False, stepwise=True, suppress_warnings=True, max_order=5)
        auto_forecast = auto_model.predict(n_periods=len(test))
        log_auto_forecast = full_log_series.iloc[train_end - 1] + np.cumsum(auto_forecast)
        auto_final = np.exp(log_auto_forecast)

        # Manual ARIMA (1,0,1)
        manual_model = ARIMA(train, order=arima_order)
        manual_fit = manual_model.fit()
        manual_forecast = manual_fit.forecast(steps=len(test))
        log_manual_forecast = full_log_series.iloc[train_end - 1] + manual_forecast.cumsum()
        manual_final = np.exp(log_manual_forecast)

        actual = original_series.iloc[train_end:test_end]

        axs[i].plot(original_series.index[:test_end], original_series[:test_end], label='Original Data', color='steelblue')
        axs[i].plot(actual.index, auto_final, label='Auto ARIMA Forecast', color='darkorange')
        axs[i].plot(actual.index, manual_final, label='Manual ARIMA (1,0,1)', color='forestgreen')
        axs[i].set_title(f"Backtest Forecast: {int(split*100)}% to {int((split+steps_fraction)*100)}%")
        axs[i].legend()
        axs[i].grid(True)

    # Future Forecast (next 10–15%)
    steps_ahead = int(future_fraction * n)

    # Auto ARIMA on full data
    auto_model = auto_arima(series, seasonal=False, stepwise=True, suppress_warnings=True, max_order=5)
    auto_forecast = auto_model.predict(n_periods=steps_ahead)
    log_auto_forecast = full_log_series.iloc[-1] + np.cumsum(auto_forecast)
    auto_final = np.exp(log_auto_forecast)

    # Manual ARIMA (1,0,1) on full data
    manual_model = ARIMA(series, order=arima_order)
    manual_fit = manual_model.fit()
    manual_forecast = manual_fit.forecast(steps=steps_ahead)
    log_manual_forecast = full_log_series.iloc[-1] + manual_forecast.cumsum()
    manual_final = np.exp(log_manual_forecast)

    # Future dates
    last_date = original_series.index[-1]
    future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=steps_ahead, freq='B')

    axs[-1].plot(original_series, label='Historical Data', color='steelblue')
    axs[-1].plot(future_dates, auto_final, label='Auto ARIMA Forecast', color='darkorange')
    axs[-1].plot(future_dates, manual_final, label='Manual ARIMA (1,0,1)', color='forestgreen')
    axs[-1].set_title("Future Forecast (Next 10–15%)")
    axs[-1].legend()
    axs[-1].grid(True)

    plt.tight_layout()
    plt.show()
compare_forecasts(
    series=log_diff_ts,
    original_series=original_ts,
    full_log_series=log_ts,
    splits=[0.2, 0.4, 0.6, 0.8],
    steps_fraction=0.2,
    arima_order=(2, 0, 0),
    future_fraction=0.15
)
```

```python
import warnings
from statsmodels.tsa.arima.model import ARIMA

def evaluate_arima_orders(series, p_range=3, d_range=2, q_range=3):
    results = []
    warnings.filterwarnings("ignore")

    for p in range(p_range + 1):
        for d in range(d_range + 1):
            for q in range(q_range + 1):
                try:
                    model = ARIMA(series, order=(p, d, q))
                    result = model.fit()
                    results.append({
                        "order": (p, d, q),
                        "aic": result.aic,
                        "bic": result.bic
                    })
                except:
                    continue

    df_results = pd.DataFrame(results).sort_values(by="aic")
    return df_results
evaluate_arima_orders(log_diff_ts)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_residual_diagnostics(residuals, title="Residual Diagnostics"):
    fig, axs = plt.subplots(1, 2, figsize=(12, 4))

    # Histogram
    axs[0].hist(residuals, bins=20, color='skyblue', edgecolor='black')
    axs[0].set_title("Histogram of Residuals")

    # Q-Q Plot
    stats.probplot(residuals, dist="norm", plot=axs[1])
    axs[1].set_title("Q-Q Plot")

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
model = ARIMA(log_diff_ts, order=(1,0,1)).fit()
plot_residual_diagnostics(model.resid)
```

```python
def diagnose_residuals(residuals):
    from statsmodels.stats.diagnostic import acorr_ljungbox
    from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
    import matplotlib.pyplot as plt

    print("Ljung-Box Test (lags=10):")
    print(acorr_ljungbox(residuals, lags=[10], return_df=True))

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plot_acf(residuals, ax=plt.gca(), lags=20)
    plt.title("Residual ACF")

    plt.subplot(1, 2, 2)
    plot_pacf(residuals, ax=plt.gca(), lags=20)
    plt.title("Residual PACF")

    plt.tight_layout()
    plt.show()
```
