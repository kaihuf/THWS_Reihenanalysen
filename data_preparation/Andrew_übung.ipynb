```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller, kpss

# -------------------------------
# 1. CSV-Daten laden
# -------------------------------
df = pd.read_csv("StockData_Samsung_final.csv")
df["Date"] = pd.to_datetime(df["Date"])
ts = df["Adj Close"].dropna()

# -------------------------------
# 2. Hilfsfunktion für Tests & Plots
# -------------------------------
def stationarity_tests(ts_data, title="Zeitreihe", dates=None):
    print(f"\n=== {title} ===")

    # Zeitverlauf plotten (anstelle Histogramm)
    plt.figure(figsize=(10, 4))
    if dates is not None:
        plt.plot(dates, ts_data, color='steelblue')
    else:
        plt.plot(ts_data, color='steelblue')
    plt.title(f"Zeitreihe – {title}")
    plt.xlabel("Datum")
    plt.ylabel("Wert")
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Histogramm anzeigen
    plt.figure(figsize=(6, 4))
    sns.histplot(ts_data, kde=True, bins=30, color='skyblue')
    plt.title(f"Histogramm – {title}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ADF-Test
    adf_result = adfuller(ts_data, autolag="AIC")
    print("→ ADF-Test:")
    print(f"  Teststatistik: {adf_result[0]:.4f}")
    print(f"  p-Wert: {adf_result[1]:.4f}")
    print(f"  Kritische Werte: {adf_result[4]}")
    if adf_result[1] < 0.05:
        print("  ✅ ADF: Stationär (H0 abgelehnt)")
    else:
        print("  ❌ ADF: Nicht stationär (H0 nicht abgelehnt)")

    # KPSS-Test
    kpss_result = kpss(ts_data, regression='c', nlags='auto')
    print("\n→ KPSS-Test:")
    print(f"  Teststatistik: {kpss_result[0]:.4f}")
    print(f"  p-Wert: {kpss_result[1]:.4f}")
    print(f"  Kritische Werte: {kpss_result[3]}")
    if kpss_result[1] < 0.05:
        print("  ❌ KPSS: Nicht stationär (H0 abgelehnt)")
    else:
        print("  ✅ KPSS: Stationär (H0 nicht abgelehnt)")

# -------------------------------
# 3. Originaldaten analysieren
# -------------------------------
stationarity_tests(ts, title="Originale Zeitreihe", dates=df["Date"])

# -------------------------------
# 4. Log-Transformation
# -------------------------------
log_ts = np.log(ts)
stationarity_tests(log_ts, title="Log-transformierte Zeitreihe", dates=df["Date"])

# -------------------------------
# 5. Differenzierung der Originaldaten
# -------------------------------
diff1 = ts.diff().dropna()
stationarity_tests(diff1, title="1. Differenz der Originaldaten", dates=df["Date"].iloc[1:])

diff2 = diff1.diff().dropna()
stationarity_tests(diff2, title="2. Differenz der Originaldaten", dates=df["Date"].iloc[2:])

# -------------------------------
# 6. Differenzierung der log-transformierten Daten
# -------------------------------
log_diff1 = log_ts.diff().dropna()
stationarity_tests(log_diff1, title="1. Differenz der log-transformierten Reihe", dates=df["Date"].iloc[1:])

log_diff2 = log_diff1.diff().dropna()
stationarity_tests(log_diff2, title="2. Differenz der log-transformierten Reihe", dates=df["Date"].iloc[2:])
```
